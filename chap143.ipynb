## Geometric Transformations of Images

##### Scaling

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('rabbit.jpg')

res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
cv2.imwrite('rerabbit.jpg',res)
```

```python
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
cv2.imwrite('rerabbit1.jpg',res)
```

##### Translation

```python
import cv2
import numpy as np

img = cv2.imread('dsu1.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Rotation

```python
img = cv2.imread('dsu6.jpg',0)
rows,cols = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),30,1)
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Affine Transformation

```python
from matplotlib import pyplot as plt

img = cv2.imread('dsu6.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,70],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

##### Perspective Transformation

```python
img = cv2.imread('dsu1.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[327,278],[759,314],[327,539],[768,512]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
